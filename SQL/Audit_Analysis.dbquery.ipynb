```sql
SELECT DATE_YEAR, FIRM_NAME, INDUSTRY_AFFECTED, SUM(TOTAL_AUDIT_Engagements)AS YEARLY_INDUSTRY_AUDIT FROM big_4_staging
GROUP BY DATE_YEAR, FIRM_NAME, INDUSTRY_AFFECTED
ORDER BY date_Year,Industry_Affected, YEARLY_INDUSTRY_AUDIT DESC;

SELECT FIRM_NAME,INDUSTRY_AFFECTED, ROUND(AVG(TOTAL_AUDIT_Engagements),1) AS YEARLY_AVG_INDUSTRY_AUDIT FROM big_4_staging
GROUP BY FIRM_NAME, INDUSTRY_AFFECTED
ORDER BY INDUSTRY_AFFECTED, YEARLY_AVG_INDUSTRY_AUDIT DESC;

SELECT DATE_YEAR, FIRM_NAME, INDUSTRY_AFFECTED,SUM(HIGH_RISK_CASES) AS YEARLY_HIGH_RISK_CASES FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR , INDUSTRY_AFFECTED
ORDER BY date_Year, INDUSTRY_AFFECTED, YEARLY_HIGH_RISK_CASES DESC;

SELECT FIRM_NAME, INDUSTRY_AFFECTED, ROUND(AVG(HIGH_RISK_CASES),1) AS AVG_HIGH_RISK_CASES FROM big_4_staging
GROUP BY FIRM_NAME, INDUSTRY_AFFECTED
ORDER BY INDUSTRY_AFFECTED, AVG_HIGH_RISK_CASES DESC;

SELECT DATE_YEAR, FIRM_NAME,INDUSTRY_AFFECTED, SUM(COMPLIANCE_VIOLATIONS) AS YEARLY_COMPLIANCE_VIOLATIONS FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR, Industry_Affected
ORDER BY date_Year, Industry_Affected, YEARLY_compliance_VIOLATIONS DESC;

SELECT FIRM_NAME, INDUSTRY_AFFECTED, ROUND(AVG(COMPLIANCE_VIOLATIONS),1) AS AVG_COMPLIANCE_VIOLATIONS FROM big_4_staging
GROUP BY FIRM_NAME, INDUSTRY_AFFECTED
ORDER BY INDUSTRY_AFFECTED, AVG_COMPLIANCE_VIOLATIONS DESC;

SELECT DATE_YEAR, FIRM_NAME,INDUSTRY_AFFECTED, SUM(FRAUD_CASES_DETECTED) AS YEARLY_FRAUD_CASES FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR, Industry_Affected
ORDER BY date_Year,Industry_Affected, YEARLY_FRAUD_CASES DESC;

SELECT FIRM_NAME, INDUSTRY_AFFECTED, ROUND(AVG(FRAUD_CASES_DETECTED),1) AS AVG_FRAUD_CASES FROM big_4_staging
GROUP BY FIRM_NAME, Industry_Affected
ORDER BY Industry_Affected, AVG_FRAUD_CASES DESC;

SELECT DATE_YEAR, FIRM_NAME, INDUSTRY_AFFECTED, ROUND(SUM(TOTAL_REVENUE_IMPACT),1) AS SUM_REVENUE_IMPACT FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR, Industry_Affected
ORDER BY date_Year, Industry_Affected, sum_revenue_IMPACT DESC;

SELECT FIRM_NAME, INDUSTRY_AFFECTED, ROUND(AVG(TOTAL_REVENUE_IMPACT),1) AS AVG_REVENUE_IMPACT FROM big_4_staging
GROUP BY FIRM_NAME, Industry_Affected
ORDER BY Industry_Affected, AVG_REVENUE_IMPACT DESC;
```
