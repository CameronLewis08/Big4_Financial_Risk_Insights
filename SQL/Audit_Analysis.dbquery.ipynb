```sql
SELECT DATE_YEAR, FIRM_NAME, INDUSTRY_AFFECTED, SUM(TOTAL_AUDIT_Engagements)AS YEARLY_INDUSTRY_AUDIT FROM big_4_staging
GROUP BY DATE_YEAR, FIRM_NAME, INDUSTRY_AFFECTED
ORDER BY date_Year,Industry_Affected, YEARLY_INDUSTRY_AUDIT DESC;

SELECT FIRM_NAME,INDUSTRY_AFFECTED, ROUND(AVG(TOTAL_AUDIT_Engagements),1) AS YEARLY_AVG_INDUSTRY_AUDIT FROM big_4_staging
GROUP BY FIRM_NAME, INDUSTRY_AFFECTED
ORDER BY INDUSTRY_AFFECTED, YEARLY_AVG_INDUSTRY_AUDIT DESC;

SELECT DATE_YEAR, FIRM_NAME, INDUSTRY_AFFECTED,SUM(HIGH_RISK_CASES) AS SUM_HIGH_RISK FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR , INDUSTRY_AFFECTED
ORDER BY date_Year, INDUSTRY_AFFECTED, sum_HIGH_risk DESC;

SELECT DATE_YEAR, FIRM_NAME, INDUSTRY_AFFECTED, AVG(HIGH_RISK_CASES) AS AVG_HIGH_RISK FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, AVG_HIGH_RISK DESC;

SELECT DATE_YEAR, FIRM_NAME,SUM(COMPLIANCE_VIOLATIONS) AS SUM_COMPLIANCE FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, sum_compliance DESC;

SELECT DATE_YEAR, FIRM_NAME, AVG(COMPLIANCE_VIOLATIONS) AS AVG_COMPLIANCE FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, AVG_COMPLIANCE DESC;

SELECT DATE_YEAR, FIRM_NAME,SUM(FRAUD_CASES_DETECTED) AS SUM_FRAUD FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, sum_fraud DESC;

SELECT DATE_YEAR, FIRM_NAME, AVG(FRAUD_CASES_DETECTED) AS AVG_FRAUD FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, AVG_FRAUD DESC;

SELECT DATE_YEAR, FIRM_NAME,SUM(TOTAL_REVENUE_IMPACT) AS SUM_REVENUE FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, sum_revenue DESC;

SELECT DATE_YEAR, FIRM_NAME, AVG(TOTAL_REVENUE_IMPACT) AS AVG_REVENUE FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, AVG_REVENUE DESC;
```
