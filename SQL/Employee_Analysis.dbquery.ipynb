```sql
SELECT FIRM_NAME, ROUND(AVG(Employee_Workload)*52,1) AS AVG_YEARLY_WORKLOAD FROM big_4_staging
GROUP BY FIRM_NAME
ORDER BY AVG_YEARLY_workload DESC ;

SELECT FIRM_NAME, INDUSTRY_AFFECTED, ROUND(AVG(Employee_Workload)*52,1) AS AVG_YEARLY_WORKLOAD FROM big_4_staging
GROUP BY FIRM_NAME , Industry_Affected
ORDER BY  Industry_Affected, AVG_YEARLY_workload DESC ;

SELECT DATE_YEAR, FIRM_NAME, Industry_Affected, ROUND(SUM(Employee_Workload)*52,1) AS TOTAL_YEARLY_WORKLOAD FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR, Industry_Affected
ORDER BY date_Year, Industry_Affected, TOTAL_YEARLY_workload DESC ;
```
