```sql
SELECT DATE_YEAR, FIRM_NAME, ROUND(AVG(Employee_Workload)*52,1) AS AVG_YEARLY_WORKLOAD FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, AVG_YEARLY_workload DESC ;

SELECT DATE_YEAR, FIRM_NAME, ROUND(SUM(Employee_Workload)*52,1) AS TOTAL_YEARLY_WORKLOAD FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, TOTAL_YEARLY_workload DESC ;

SELECT FIRM_NAME, ROUND(SUM(Employee_Workload)*52,1) AS TOTAL_WORKLOAD FROM big_4_staging
GROUP BY FIRM_NAME
ORDER BY TOTAL_workload DESC ;
```
