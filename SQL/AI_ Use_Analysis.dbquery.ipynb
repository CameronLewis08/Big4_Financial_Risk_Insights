```sql
SELECT DATE_YEAR, FIRM_NAME, COUNT(AI_Used_for_Auditing) AS YES_AI FROM big_4_staging 
WHERE AI_Used_for_Auditing = 'Yes'
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, YES_AI DESC;

WITH AI_USED AS 
(
  SELECT DATE_YEAR, FIRM_NAME, COUNT(AI_USED_FOR_AUDITING) AS AI_YES FROM big_4_staging
  WHERE AI_Used_for_Auditing = 'Yes'
  GROUP BY FIRM_NAME, DATE_YEAR
  ORDER BY date_Year, AI_YES DESC
)
SELECT FIRM_NAME,ROUND(AVG(AI_YES),1) AS AVG_YES FROM AI_USED 
  GROUP BY FIRM_NAME
  ORDER BY FIRM_NAME;

WITH AI_USED AS 
(
  SELECT DATE_YEAR, FIRM_NAME, COUNT(AI_USED_FOR_AUDITING) AS AI_YES FROM big_4_staging
  WHERE AI_Used_for_Auditing = 'Yes'
  GROUP BY FIRM_NAME, DATE_YEAR
  ORDER BY date_Year, AI_YES DESC
)
SELECT DATE_YEAR,SUM(AI_YES) AS SUM_YES FROM AI_USED 
  GROUP BY  DATE_YEAR
  ORDER BY  DATE_YEAR;


WITH AI_USED AS 
(
  SELECT DATE_YEAR, FIRM_NAME, COUNT(AI_USED_FOR_AUDITING) AS AI_NO FROM big_4_staging
  WHERE AI_Used_for_Auditing = 'No'
  GROUP BY FIRM_NAME, DATE_YEAR
  ORDER BY date_Year, AI_NO DESC
)
SELECT FIRM_NAME, ROUND(AVG(AI_NO),1) AS AVG_NO FROM AI_USED 
  GROUP BY FIRM_NAME
  ORDER BY FIRM_NAME;


WITH AI_USED AS 
(
  SELECT DATE_YEAR, FIRM_NAME, COUNT(AI_USED_FOR_AUDITING) AS AI_NO FROM big_4_staging
  WHERE AI_Used_for_Auditing = 'No'
  GROUP BY FIRM_NAME, DATE_YEAR
  ORDER BY date_Year, AI_NO DESC
)
SELECT DATE_YEAR,SUM(AI_NO) AS SUM_NO FROM AI_USED 
  GROUP BY  DATE_YEAR
  ORDER BY  DATE_YEAR;
```
