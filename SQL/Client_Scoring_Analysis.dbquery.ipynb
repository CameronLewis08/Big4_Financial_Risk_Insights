```sql

SELECT FIRM_NAME, AVG(Client_Satisfaction_Score) AS SATISFACTION FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR
ORDER BY date_Year, satisfaction DESC;

SELECT FIRM_NAME, INDUSTRY_AFFECTED, ROUND(AVG(Audit_Effectiveness_Score),1) AS AVG_YEARLY_EFFECTIVENESS FROM big_4_staging
GROUP BY FIRM_NAME, Industry_Affected
ORDER BY Industry_Affected, AVG_YEARLY_effectiveness;

SELECT DATE_YEAR, FIRM_NAME, iNDUSTRY_AFFECTED, MIN(AUDIT_EFFECTIVENESS_SCORE) AS MIN_EFFECTIVENESS FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR, iNDUSTRY_AFFECTED
ORDER BY date_Year, iNDUSTRY_AFFECTED, MIN_EFFECTIVENESS;

SELECT DATE_YEAR, FIRM_NAME, iNDUSTRY_AFFECTED, MAX(AUDIT_EFFECTIVENESS_SCORE) AS MAX_EFFECTIVENESS FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR, iNDUSTRY_AFFECTED
ORDER BY date_Year, iNDUSTRY_AFFECTED, MAX_EFFECTIVENESS;

SELECT DATE_YEAR, FIRM_NAME, iNDUSTRY_AFFECTED, MIN(CLIENT_SATISFACTION_SCORE) AS MIN_SATISFACTION FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR, iNDUSTRY_AFFECTED
ORDER BY date_Year, iNDUSTRY_AFFECTED, MIN_SATISFACTION;

SELECT DATE_YEAR, FIRM_NAME, iNDUSTRY_AFFECTED, MAX(CLIENT_SATISFACTION_SCORE) AS MAX_SATISFACTION FROM big_4_staging
GROUP BY FIRM_NAME, DATE_YEAR, iNDUSTRY_AFFECTED
ORDER BY date_Year, iNDUSTRY_AFFECTED, MAX_SATISFACTION;

SELECT FIRM_NAME, INDUSTRY_AFFECTED, ROUND(AVG(Client_Satisfaction_Score),1) AS YEARLY_AVG_SATISFACTION FROM big_4_staging
GROUP BY FIRM_NAME, Industry_Affected
ORDER BY Industry_Affected, yeaRLY_AVG_satisfaction DESC;
```
